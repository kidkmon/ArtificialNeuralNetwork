```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
```

## Atividade 7 -- Histograma

Construa o histograma do atributo alvo mpg utilizando bins de tamanho 1. O seu histograma deve ser intitulado 'Consumo', o eixo 'y' deve apresentar a unidade de medida 'milhas por galão'. O eixo x deve ser rotulado com os valores associados aos bins. A cor do gráfico deve ser em tons de cinza.

```python
dataset = pd.read_csv('autompg.csv', delimiter=';')
dataset.dropna(inplace=True)

mpg = dataset['mpg']

plt.hist(mpg, bins=range(1,int(mpg.max())), color='gray')
plt.title('Consumo')
plt.xlabel('Milhas por galão')
plt.show()
```

## Atividade 8 -- Boxplot

1. Obtenha o boxplot do atributo alvo mpg
2. Obtenha o boxplot do peso dos carros
3. Responda: o peso dos carros são distribuído de maneira simétrica?

#### Boxplot do atributo alvo mpg:

```python
dataset.boxplot(column='mpg')
plt.show()
```

#### Boxplot do peso dos carros:

```python
dataset.boxplot(column='weight')
plt.show()
```

#### O peso dos carros são distribuído de maneira simétrica?

```python
dataset['weight'].skew()
```

De acordo com o boxplot e o cálculo acima da simetria da coluna de peso do dataset, é possível dizer que o peso dos carros não são distribuídos de maneira simétrica devido ao valor da sua simetria ser maior que 0, o que resulta em uma assimetria positiva.
